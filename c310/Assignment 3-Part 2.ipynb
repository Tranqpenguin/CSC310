```python
import sys
sys.path.insert(0,'/C:/Users/andre/Documents/c310')
import treeviz
import pandas as pd
import os
path = os.path.join('c:'+ os.sep,'Users','andre','Documents','c310', 'Guns.csv')
df = pd.read_csv(path)
from sklearn import tree
from treeviz import tree_print
from sklearn.metrics import accuracy_score
```

```python
df.head()
```

```python
features_df = df.drop(['state', 'law'],axis=1)
features_df.head()
```

```python
target_df = pd.DataFrame(df['law'])
target_df.head()
```

```python
dtree = tree.DecisionTreeClassifier(criterion='entropy')
dtree.fit(features_df,target_df)
```

```python
tree_print(dtree,features_df)
```

```python
predict_array = dtree.predict(features_df)      # produces an array of labels
predicted_labels = pd.DataFrame(predict_array)  # turn it into a DF
predicted_labels.columns = ['law'] 
```

```python
predicted_labels.head()
```

```python
target_df.head()
```

```python
predicted_labels.equals(target_df)
```

```python
print("Our model accuracy is: {}".format(accuracy_score(target_df, predicted_labels)))
```
